```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
dtypes = {
    'title': 'str',
    'year': 'str',
    'stars': 'str',
    'link': 'str'
}

df = pd.read_csv('dataset.csv', dtype=dtypes)
df.head()
```

```python
df['year'] = df.year.replace('-', 0)
df['year'] = df.year.astype('int')
```

```python
df['avg_stars'] = df.stars.apply(lambda x: np.mean(eval(x)) if not x=='[nan]' else 0)
```

```python
df.dtypes
```

```python
df.avg_stars.describe()
```

```python
plt.figure(figsize=(16, 6))
ax = sns.barplot(x="year", y="avg_stars", data=df.query("year >= 1900"))
plt.title("Average Stars per Release Year")
plt.ylabel('Avg. Stars');
```

```python
sns.jointplot(x="year", y="avg_stars", data=df.query("year >= 1900"), kind='kde', height=8);
```

```python
sns.relplot(x="year", y="avg_stars", data=df.query("year >= 1900"), height=8);
```
